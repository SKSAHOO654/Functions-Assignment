**1- What is the difference between a function and a method in Python?**

1.Function:
- A function is an independent block of reusable code.
- It is defined using the def keyword.
- It can be called directly without being associated with any object.
- Functions can exist outside of classes.


2.Method:
- A method is a function that is associated with an object (i.e., belongs to a class).
- It is defined inside a class and operates on the instance of the class.
- Methods must be called on an object (instance of a class).
- The first parameter of an instance method is usually self, which represents the object itself.


**2.Explain the concept of function arguments and parameters in Python.**

In Python, parameters and arguments are terms used in the context of functions:
- Parameters: Variables listed in the function definition.
- Arguments: Values passed to a function when it is called.

1.Positional Arguments
- These are arguments passed in the correct order as defined in the function.
- Their position matters.

2.Default Arguments
- Parameters can have default values.
- If no argument is passed, the default value is used.

3.Keyword Arguments (Named Arguments)
- Arguments can be passed using parameter names (keywords).
- The order does not matter.

4.Variable-Length Arguments (*args and **kwargs)
- *args (Non-Keyword Variable-Length Arguments)
  - Allows passing multiple values as a tuple.
- **kwargs (Keyword Variable-Length Arguments)
  - Allows passing multiple keyword arguments as a dictionary.

5.Positional-Only and Keyword-Only Arguments
- Positional-only arguments (/): Must be passed by position.
- Keyword-only arguments (*): Must be passed by name.

10.Function with map(), filter(), and reduce()

**4. What is the purpose of the `return` statement in a Python function?**

3-Can Return Multiple Values
4-Can Return Any Data Type

**6. Explain the concept of generators in Python and how they are defined.**

- A generator in Python is a special type of iterable that produces values lazily (on demand) using the yield keyword instead of returning all values at once.
 - Memory Efficient – Generators don’t store all values in memory, making them ideal for large datasets.
 - Stateful – They remember where they left off, unlike normal functions.
 - Lazy Evaluation – Values are generated only when needed, reducing computation overhead.

1.Using Generators in a for Loop
- Instead of manually calling next(), we can iterate using a for loop.

2.Generator Expression (Shorter Syntax)
- Instead of defining a function, we can create generators using generator expressions, similar to list comprehensions.

3.Generator Use Cases
- Generators are useful when dealing with large datasets or infinite sequences.

4.Chaining Generators
- We can chain generators together for better modularity.

**7. What are the advantages of using generators over regular functions?**

- Generators provide several key advantages over regular functions,
especially when dealing with large data sets or streams of data. Here’s why generators are often preferred:

1.Memory Efficiency
 - Generators don’t store all values in memory; they produce values on demand.
 - Regular functions return entire lists, consuming more memory.
  - Example: Large List vs. Generator
 - The list function stores all values in memory at once, while the generator produces one value at a time.

2.Faster Execution
- Generators improve performance by avoiding unnecessary computations.
  - Example: Processing Large Data
- Generators execute faster because they compute values lazily.

3.Infinite Sequences
- Generators can generate infinite sequences without consuming memory, which is impossible with regular functions.
 - Example: Infinite Fibonacci Generator
- Regular functions can’t handle infinite sequences, but generators can!

4.Lazy Evaluation
- Generators compute values only when needed, which is useful for large datasets.
 - Example: Filtering Even Numbers
- Saves computation by only generating required values

5.Simpler Code & Better Readability
- Using yield avoids creating large lists and makes the code simpler.
 - Example: Normal Function vs. Generator
- Generators use fewer lines of code and are easier to read.

6.Generators Can Be Chained
- You can combine multiple generators efficiently.
 - Example: Chaining Generators
- Chaining allows modular and efficient processing.

7.Avoids Recursion Depth Limits
- Generators avoid the RecursionError in deeply recursive functions.
 - Example: Generator for Factorial Instead of Recursion
- Regular recursion has depth limits, but generators avoid this problem.

8.Generators Work Well With Streams
- Generators process streams of data efficiently, making them useful for:
 - ✔ Reading large files line by line
 - ✔ Processing live data from APIs
   - Example: Reading a Large File Line-by-Line
- Generates each line on demand, avoiding memory overload.


**8. What is a lambda function in Python and when is it typically used?**

- A lambda function in Python is a small, anonymous function that can have any number of arguments but only one expression. It is created using the lambda keyword.

  - Lambda functions are typically used when:
    - 1-A function is needed temporarily (i.e., without defining a full function).
    - 2-Used as an argument in functions like map(), filter(), and sorted().
    - 3-For quick inline operations in list comprehensions or one-time calculations.

**9. Explain the purpose and usage of the `map()` function in Python.**

- The map() function in Python is used to apply a function to each item in an iterable (such as a list, tuple, or set) and
 returns a new iterable (a map object). It helps in functional programming by eliminating the need for explicit loops.

  - Usage of map()
    - 1: Convert a list of numbers to their squares
    - 2: Convert a list of temperatures from Celsius to Fahrenheit
    - 3: Using map() with Multiple Iterables

- Benefits of map()
 -  Concise & Readable – Eliminates the need for explicit loops.
 - Efficient – Works faster than loops, especially for large datasets.
 -  Functional Programming – Works well with lambda functions.
